# 處理股價

```python
import pandas as pd
import datetime
```

### 讀取資料

```python
stock = pd.read_excel(r"stock_data_2019-2021.xlsx", sheet_name=["上市2019", "上市2020", "上市2021"])
stock = pd.concat([stock["上市2019"], stock["上市2020"], stock["上市2021"]], axis = 0)
stock
```

將日期與代碼獨立成一欄 並篩選掉

```python
def code(string):
    return string.split()[0]
stock["code"] = stock["證券代碼"].apply(lambda x:code(x))
stock['day'] = stock["年月日"]
stock = stock.replace("-", None)
stock
```

將資料轉換為 index = 公司代號, columns為日期+價格

```python
stock_pivot = stock.pivot(index = "code", columns = 'day', values=["開盤價(元)", "收盤價(元)"])
date = []
for i in stock_pivot.columns.values:
    date.append(i[1])
stock_pivot.columns = ['_'.join(col[::-1]) for col in stock_pivot.columns.values]
stock_pivot.head()
```

### 計算股價變動 

股價變動為：新聞發布後第三日之收盤價 - 隔一日之開盤價

```python
stock_delta = pd.DataFrame(index = stock_pivot.index)
for day in date:
    nextday = (datetime.datetime.strptime(day, '%Y/%m/%d') + datetime.timedelta(days = 1))
    thirdday = nextday + datetime.timedelta(days = 2)
    nextday = nextday.strftime('%Y/%m/%d')
    thirdday = thirdday.strftime('%Y/%m/%d')
    if nextday in date and thirdday in date:
        stock_delta[day] = stock_pivot[thirdday+'_收盤價(元)'] - stock_pivot[nextday+'_開盤價(元)']
    else:
        stock_delta[day] = None
```

變動數>0 = 漲
變動數<0 = 跌

```python
stock_delta = stock_delta.dropna(axis = 1, how = "all")
def upordown(delta):
    if delta>0:
        return '漲'
    elif delta <0:
        return '跌'
    else:
        return delta
for day in stock_delta.columns:
    stock_delta[day] = stock_delta[day].apply(lambda x: upordown(x) )
```

```python
stock_delta
```

存檔

```python
stock_delta.to_pickle('stock_delta.pickle')
```

### 標記公司

```python
# 讀取所有上市公司之資料
company_information = pd.read_csv('company_information.csv')
company_information["公司代號"] = company_information["公司代號"].apply(lambda x: str(x))
company_information.head()
```

將公司名稱、簡稱、董事長當作個股的關鍵字

```python

company_word = pd.merge(company_information[['公司名稱', '公司簡稱', '董事長', '公司代號']], pd.DataFrame(stock.index), how = 'left', left_on="公司代號", right_on = 'code')
company_word = company_word.dropna(axis = 0)
company_word = company_word.drop("公司代號", axis = 1)
company_word = company_word.set_index("code")
```

```python
word = {}
for com in company_word.index:
    inf = list(company_word.loc[com])
    # word[com] = inf[0] + "|" + inf[1] +"|" + inf[2]
    word[com] = inf
```

標記文章


```python
def mark_company(string):
    com_list = []
    for com in word:
        wo = word[com]
        if wo[0] in string or wo[1] in string or wo[2] in string:
            com_list.append(com)
    return com_list
```

```python
allnews['company'] = allnews['full_content'].apply(lambda x:mark_company(x))
```

```python
# 新增一欄位為company為此新聞包含的所有公司代碼
allnews.to_pickle('allnews_with_company.pickle')
```

由於一則新聞與超過一個個股相關可能造成混亂，因此只篩選出對應到一個個股的新聞

```python
company_news = allnews[allnews['company'].apply(lambda x:len(x))==1].copy()
company_news['code'] = company_news['company'].apply(lambda x: x[0])
company_news
```

### 標記漲跌

```python
# 示範使用melt轉換格式
stock_delta = stock.copy()
stock_delta.reset_index(inplace = True)
stock_delta.melt('code', var_name= 'date', value_name='delta')
# pd.melt(stock_delta, id_vars='code', var_name = 'date', value_name = 'delta')
```

```python
# 將股票代號、日期作為合併的依據，將文章以及漲跌做合併
company_news_delta = pd.merge(company_news, stock_delta.melt('code', var_name= 'date', value_name='delta'), how = 'left', on=['date', 'code'])
company_news_delta = company_news_delta.dropna(axis = 0, how = 'any')
company_news_delta.drop('company', inplace = True, axis = 1) # 有些新聞當日沒有漲跌的資料
```

```python
company_news_delta = company_news_delta.reset_index()
company_news_delta.drop('index', axis = 1, inplace=True)
company_news_delta.to_pickle('newsall_company_delta.pickle')
company_news_delta.head()
```
